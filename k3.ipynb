```python
import numpy as np

# Define the function funny(x)
def funny(x):
    if x <= -1:
        return 0
    elif x >= 1:
        return 0
    else:
        return np.exp(1 - (1 / (1 - x ** (10 ** 250))))

# (a) Check if funny(x) is continuous at x = ±1
def is_continuous():
    # Check if the function approaches the same value from the left and right sides of x = ±1
    if funny(1) == funny(1) == 0:
        return "The function is continuous at x = ±1."
    else:
        return "The function is not continuous at x = ±1."

# (b) Evaluate limx→1− f(x) using a substitution
def limit_at_1_minus():
    # Substitute y = -1 / (1 - x^(10^250))
    # As x approaches 1 from the left, y approaches -∞
    # Therefore, the limit as y approaches -∞ is +∞
    return "+∞"


# (4a) Implementing the bisection method to find roots of a function
def roots(f, a, b):
    # Check if f(a) and f(b) have opposite signs
    if f(a) * f(b) > 0:
        return "No root found in the given interval."

    # Iterate until the desired precision is reached
    precision = 10 ** -10
    while abs(b - a) > precision:
        c = (a + b) / 2
        if f(c) == 0:
            return round(c, 10)  # Return the root with 10 decimal places
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return round(c, 10)

# Test cases for the roots function
def test_roots():
    # Test case 1: f(x) = e^x + ln(x) on [0, 1]
    def f1(x):
        return np.exp(x) + np.log(x)
    print("Root of f1(x) =", roots(f1, 0.1, 1))

    # Test case 2: f(x) = arctan(x) - x^2 on [0, 2]
    def f2(x):
        return np.arctan(x) - x ** 2
    print("Root of f2(x) =", roots(f2, 0, 2))

    # Test case 3: f(x) = sin(x) / ln(x) on [3, 4]
    def f3(x):
        return np.sin(x) / np.log(x)
    print("Root of f3(x) =", roots(f3, 3, 4))

    # Test case 4: f(x) = ln(cos(x)) on [5, 7]
    def f4(x):
        return np.log(np.cos(x))
    print("Root of f4(x) =", roots(f4, 5, 7))

# Calling the functions to demonstrate
print("(a) Is funny(x) continuous?")
print(is_continuous())
print()

print("(b) Evaluate lim(x→1-) funny(x):")
print("lim(x→1-) funny(x) =", limit_at_1_minus())
print()

print("(4a) Finding roots of functions:")
test_roots()
```
