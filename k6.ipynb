```python
import numpy as np
import matplotlib.pyplot as plt

def normal_density(mean, variance, x):
    # Calculate the standard deviation from the variance
    std_dev = np.sqrt(variance)
    
    # Calculate the density function value using the formula
    density = 1 / (np.sqrt(2 * np.pi) * std_dev) * np.exp(-((x - mean) ** 2) / (2 * variance))
    
    return density

def plot_normal_density(mean, variance, x_values):
    # Calculate the density function values for the given x values
    densities = [normal_density(mean, variance, x) for x in x_values]
    
    # Plot the normal density function
    plt.plot(x_values, densities, label=f'mean={mean}, variance={variance}')
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.title('Normal Density Function')
    plt.legend()
    plt.grid(True)
    plt.show()

def numerical_integration(mean, variance, a, b, num_points=1000):
    # Define the function to be integrated
    f = lambda x: normal_density(mean, variance, x)
    
    # Calculate the width of each subdivision
    dx = (b - a) / num_points
    
    # Initialize the integral value
    integral = 0
    
    # Perform numerical integration using the midpoint rule
    for i in range(num_points):
        x_left = a + i * dx
        x_right = a + (i + 1) * dx
        integral += f((x_left + x_right) / 2) * dx
        
    return integral

# Example usage
mean_height = 171  # in cm
variance_height = 7.1 ** 2  # variance to standard deviation squared

# Plot the normal density function for different values of µ and σ^2
x_values = np.linspace(150, 200, 1000)
plot_normal_density(mean_height, variance_height, x_values)

# Calculate the probability that a randomly chosen man has a height between 162cm and 190cm
a = 162
b = 190
probability = numerical_integration(mean_height, variance_height, a, b)
print("Probability:", probability)
```
