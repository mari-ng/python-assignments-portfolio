```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(f, df, x0, learning_rate, max_iterations=1000, tolerance=1e-6):
    x_values = [x0]  # List to store the sequence of points
    iterations = 0
    
    while iterations < max_iterations:
        # Compute the derivative at the current point
        derivative = df(x_values[-1])
        
        # Update the current point using gradient descent
        x_new = x_values[-1] - learning_rate * derivative
        
        # Append the new point to the sequence
        x_values.append(x_new)
        
        # Check for convergence
        if abs(derivative) < tolerance:
            break
        
        iterations += 1
    
    return x_values

def plot_gradient_descent(f, x_values):
    # Plot the input function
    x = np.linspace(min(x_values) - 1, max(x_values) + 1, 1000)
    y = f(x)
    plt.plot(x, y, label='f(x)')
    
    # Plot the sequence of points generated by gradient descent
    plt.scatter(x_values, f(np.array(x_values)), color='red', label='Gradient Descent')
    
    # Mark the starting point
    plt.scatter(x_values[0], f(x_values[0]), color='green', label='Starting Point')
    
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gradient Descent in One Dimension')
    plt.legend()
    plt.grid(True)
    plt.show()

# Test functions
def f1(x):
    return x ** 2

def df1(x):
    return 2 * x

def f2(x):
    return -3 * x ** 3 + 10 * x ** 2 - 5 * x - 3

def df2(x):
    return -9 * x ** 2 + 20 * x - 5

def funny_function(x):
    if x > 0:
        return x ** x
    elif x == 0:
        return 1
    else:
        return abs(x) ** abs(x)

def df_funny_function(x):
    if x > 0:
        return x ** x * (np.log(x) + 1)
    elif x == 0:
        return 0
    else:
        return -abs(x) ** abs(x) * (np.log(abs(x)) + 1)

# Test cases
learning_rate = 0.1
x0 = 5

print("Test Case 1:")
x_values = gradient_descent(f1, df1, x0, learning_rate)
plot_gradient_descent(f1, x_values)

print("Test Case 2:")
x_values = gradient_descent(f2, df2, x0, learning_rate)
plot_gradient_descent(f2, x_values)

print("Test Case 3:")
x_values = gradient_descent(funny_function, df_funny_function, x0, learning_rate)
plot_gradient_descent(funny_function, x_values)
```
